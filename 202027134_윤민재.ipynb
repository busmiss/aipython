```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('iris.csv')
df.info()
print()
basic_stat = df.describe()
species_count = df['species'].value_counts()
print(basic_stat,'\n')
print(species_count)
```

```python
NA = df.isna().sum()
print(NA,'\n')
print('중복값 개수 : ',df.duplicated().sum())
print('실제 중복데이터 확인')
print(df[df.duplicated(keep = False)])
df_new = df.drop_duplicates()
```

```python
spec_sum = df_new.groupby('species').sum()
print('품종별 데이터 합계')
print(spec_sum)
spec_avg = df_new.groupby('species').mean()
print('\n품종별 데이터 평균')
print(spec_avg)
```

```python
spec_avg["sepal_length"].plot(kind='bar')
```
