```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

iris = pd.read_csv('iris.csv')

# iris=sns.load_dataset("iris")
sns.scatterplot(x='petal_length',y='petal_width'
,hue='species',data=iris)
plt.title('Scatter Plot by seaborn', fontsize=20)
plt.show()
```

```python
sns.countplot(x='species',data=iris)
plt.show()
```

```python
sns.barplot(x='species',y='petal_length',data=iris,errorbar='sd')
plt.show()
```

```python
sns.pairplot(iris,hue='species')
plt.show()
```

```python
sns.regplot(x='petal_length',y='petal_width'
, data=iris)
```

```python
df = iris.corr()

sns.heatmap(data=df,annot=True,fmt='.2f',cbar=False)
plt.show()
# 데이터를 펼쳐놓고 상관관계가 높은 정도를 산전도로 시각화
```

```python
# MISSION 3
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

penguin = pd.read_csv('penguins.csv')
sns.barplot(x='species',y='body_mass_g',data=penguin,errorbar='sd',hue='sex')
plt.show()

df = penguin.corr()
sns.heatmap(data=df,annot=True,fmt='.2f',cbar=False)
plt.show()

sns.regplot(x='body_mass_g',y='flipper_length_mm'
, data=penguin)
plt.show()
```
